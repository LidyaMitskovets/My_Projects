# Анализ динамики аренд велосипедов

### Описание данных
- timestamp – дата и время (точность до часа)
- cnt – количество аренд велосипедов за этот час
- t1 – температура, в С
- t2 – температура "ощущается как", в С 
- hum – влажность (%)
- wind_speed – скорость ветра, км/ч
- weather_code – погодные условия:
    - 1 – ясно (SKC)
    - 2 – преимущественно ясно / встречаются отдельные облака (SCT)
    - 3 – облачно / значительные облака (BKN)
    - 4 – пасмурно (OVC)
    - 7 – небольшой дождь Rain/ light Rain shower/ Light rain 
    - 10 – дождь с грозой
    - 26 – снегопад 
    - 94 – ледяной туман (да, такое бывает!)
- isholiday – является ли день праздником (1 – праздник, 0 – нет)
- isweekend –  является ли день выходным (1 – выходной, 0 – нет)
- season – метеорологический сезон (0 – весна, 1 – лето, 2 – осень, 3 – зима)

```python
# импорт библиотек
import pandas as pd
import numpy as np
import seaborn as sns 
%matplotlib inline
```

```python
# чтение данных
london = pd.read_csv('london.csv', parse_dates = [0])
london
```

```python
# проверка данных на наличие пустых значений
london.isna().sum()
```

```python
# график зависимости числа поездок от времени
sns.lineplot( data = london , x = 'timestamp', y = 'cnt')
```

Данные нечитабельны, требуется дальшейшее преобразование для определения тренда

```python
# преобразуем колонку timestamp в индексы, для удобства дальнейшей визуализации
london = london.sort_values(by = 'timestamp').set_index('timestamp')
```

```python
# визуализируем данные
london.resample(rule='D').cnt.sum()
london_agg = london.resample(rule='D').agg({'cnt':'sum'})
london_agg.plot()
```

### Вывод по графику:
- Зимой число аренд меньше, чем летом
- Максимальное число аренд за день составляет более 70 000
- Встречаются как очень резкие скачки числа аренд, так и падения (!!!)

Поиск аномалий, которые могли бы привести к резким скачкам 

```python
# чтобы сгладить ряд, используем скользящее среднее
london_agg_1 = london_agg.rolling(window = 3).mean().rename(columns = {'cnt': 'cnt_rolling'})
london_agg_1['2015-07-09':]
```

```python
london_agg_1
```

```python
london_agg
```

```python
# соединим данные о скользящем среднем и обычным средним
london_agg_all = london_agg.merge(london_agg_1, how = 'right', on = 'timestamp')
```

```python
# вычслим разницу между наблюдаемыми и сглаженными значениями
london_agg_all['dif'] = london_agg_all.cnt - london_agg_all.cnt_rolling
```

```python
london_agg_all
```

Определим верхние и нижние границы доверительного интервала

```python
# вычислим стандартное отклонение
std = np.std(london_agg_all.dif)
```

```python
std.round(2)
```

```python
london_agg_all['upper_bound'] = london_agg_all.cnt_rolling + 2.576 * std
```

```python
london_agg_all['lower_bound'] = london_agg_all.cnt_rolling - 2.576 * std
```

```python
london_agg_all
```

```python
# значение верхней границы для последнего наблюдения
london_agg_all.loc['2017-01-03', 'upper_bound'].round(2)
```

```python
# выведим наблюдения, для которых наблюдаемые значения оказались больше верхней границы 99% доверительного интервала
london_agg_all.query('cnt > upper_bound')
```

Самый большой скачок произошел 2015-07-09, выявим его причину

```python
# анализируем данные, которые есть за эти дни 
london.query('timestamp in ["2015-07-09", "2015-08-06", "2016-09-04","2016-12-25"]')
```
